```python
#getting the date,currency and mean
test_currency = cbk [['Date','Currency','Mean','Buy','Sell']]
test_currency
```

```python
def get_buying_price(Date, Currency):
    buying = test_currency [(test_currency['Currency'] == Currency) & (test_currency['Date'] == Date)]
    return int (buying ['Buy'])
```

```python
get_buying_price('2015-02-01','KES / USHS')
```

```python
currencies =  ['US DOLLAR', 'STG POUND','EURO', 'CHINESE YUAN', 'JPY (100)' ]
profit_loss_df =pd.DataFrame()
currencies
```

```python
for currency in currencies:
    buy= get_buying_price('2015-05-01', currency)
    currency_df = cbk[(cbk['Currency'] == currency) & (cbk['Date'] >= '2015-02-01') & (cbk['Date']<= '2015-07-31')]
    #print(buy)
    
    profit_loss_df[currency]= list(currency_df['Sell']-buy)
```

```python
profit_loss_df
```

```python
for currency in currencies:
    plt.plot(currency_df['Date']. sort_values(), profit_loss_df[currency])
    
plt.title('Profit and loss of each currency')
plt.xlabel('Date')
plt.xticks(rotation = 90)
plt.ylabel('Profit/loss')
plt.legend(currencies, bbox_to_anchor =(1,0.5))
    
```

```python
new_cbk = cbk[((cbk["Currency"] == "US DOLLAR") | (cbk["Currency"] == "EURO") | (cbk["Currency"] == "CHINESE YUAN") 
           | (cbk["Currency"] == "SA RAND") | (cbk["Currency"] == "STG POUND")) & (cbk["Date"] == '2015-01-21')]
new_cbk
```

```python
# plt.bar(x=new_cbk['Currency'], height=new_cbk['Mean']) # matplotlib
plt.title('Profits in each currency')
#plt.figure(figsize = (12,9))
sns.barplot(x=new_cbk['Currency'],y=new_cbk['Mean']) # seaborn
plt.ylabel('Profits')
plt.xlabel('Currency')
```

```python
new_cbk['Proportions'] = new_cbk['Sell'] / sum(new_cbk['Sell'])
new_cbk
```

```python
labels = new_cbk["Currency"]
sizes = new_cbk["Proportions"]

plt.pie(sizes, labels=labels, autopct='%0.1f%%')
plt.title('Proportion of Currencies in the Sample')
plt.axis('equal')
plt.show()
```
